```python
import pandas as pd
# 呈现图表效果导包，引入相关配置文件
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 读取数据表格
df = pd.read_csv('【清洗后数据】Python职业招聘数据-上海.csv')
df.head()
```

```python
# 划分数据
area_info = df['区域'].value_counts().index.to_list()
area_num = df['区域'].value_counts().to_list()
exp_info = df['经验'].value_counts().index.to_list()
exp_num = df['经验'].value_counts().to_list()
cop_info = df['公司性质'].value_counts().index.to_list()
cop_num = df['公司性质'].value_counts().to_list()
study_info = df['学历'].value_counts().index.to_list()
study_num = df['学历'].value_counts().to_list()
```

```python
# 引用Github上的开源pyecharts画廊里的可视化代码，进行相关数据的可视化操作
# 官方网址：https://gallery.pyecharts.org/#/
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker

c = (
    Pie()
    .add(
        "【Python-上海】各区域职位分布",
        [
            list(z)
            for z in zip(
                area_info,
                area_num,
            )
        ],
        center=["40%", "50%"],
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="上海Python岗位招聘区域分布情况"),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="80%", orient="vertical"),
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    # 保存html文件
    .render("【Python-上海】各区域职位分布-饼状图.html")
)
# 也可以选择直接在jupyter上呈现图表效果
# 在jupyter上呈现图形效果
# c.load_javascript()
```

```python
import pyecharts.options as opts
from pyecharts.charts import Bar3D
import numpy as np

# 检查原始数据中的薪资是否有非零值
print("薪资列中非零值的数量:", df[df['薪资'] > 0]['薪资'].count())
# 将薪资列转换为更大的数据类型
# df1['薪资'] = df1['薪资'].astype('int64')

# 检查数据转换后的情况
print("薪资列的数据类型:", df['薪资'].dtype)
print("薪资列中的最大值:", df['薪资'].max())

# 数据处理：按年份和地区分组，计算每个分组的薪资总和
grouped_data = df.groupby(['区域', '学历'])['薪资'].mean().reset_index()

print(grouped_data)

# 准备3D图表所需的数据格式
data = []
years = sorted(grouped_data['区域'].unique())
regions = sorted(grouped_data['学历'].unique())

for year in years:
    for region in regions:
        # 薪资，如果该年份该地区没有数据，则使用0
        play_count = grouped_data[(grouped_data['区域'] == year) & (grouped_data['学历'] == region)]['薪资'].sum()
        # 添加所有数据
        data.append([years.index(year), regions.index(region), play_count])

# 检查生成的数据
print("生成的3D图表数据:", data)

# 创建3D柱状图
bar3d = Bar3D(init_opts=opts.InitOpts(width="1600px", height="800px"))
bar3d.add(
    series_name="薪资",
    data=data,
    xaxis3d_opts=opts.Axis3DOpts(type_="category", data=years, name="区域", axislabel_opts=opts.LabelOpts(rotate=-15, interval=0)),
    yaxis3d_opts=opts.Axis3DOpts(type_="category", data=regions, name="学历"),
    zaxis3d_opts=opts.Axis3DOpts(type_="value", name="薪资"),
)
bar3d.set_global_opts(
    visualmap_opts=opts.VisualMapOpts(max_=30000, range_color=[
                "#313695",
                "#4575b4",
                "#74add1",
                "#abd9e9",
                "#e0f3f8",
                "#ffffbf",
                "#fee090",
                "#fdae61",
                "#f46d43",
                "#d73027",
                "#a50026",
            ]),
    title_opts=opts.TitleOpts(title="薪资"),
)
# 保存图表为HTML文件
bar3d.render('【Python-上海】各区域、学历薪资一览-3D图.html')
```
